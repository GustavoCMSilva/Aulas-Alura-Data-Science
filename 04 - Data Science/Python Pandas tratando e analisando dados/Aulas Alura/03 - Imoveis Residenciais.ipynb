## Relatório de Análise 3

### Tipos de Imóveis

```python
# Usaremos este recurso para nos ajudar na seleção dos imóveis.
list(dados['tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete', 
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
#apaguei os outros itens da lista manualmente
```

```python
dados['tipo'].isin(residencial)
# Ao passarmos a listagem residecial que desejamos, 
# será verificado dentro do DataFrame se os elementos da nossa lista existem de fato
```

```python
selecao = dados['tipo'].isin(residencial)
#Atribuiremos a nossa lista de imóveis residenciais a variável selecao.
```

```python
dados_residencial = dados[selecao]
dados_residencial #aqui ficou somente os dados residenciais
```

```python
list(dados_residencial['tipo'].drop_duplicates())
# Faremos uma verificação para garantir a integridade dos dados.
# Coletaremos dados_residencial, especificamente a variável Tipo, e evocaremos o método drop_duplicates().
```

```python
dados_residencial.shape[0] 
# Checaremos o tamanho DataFrame por meio de shape().
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
# O próximo passo é realizar reconstrução do índice de dados_residencial.
dados_residencial
```

### Exportando/Salvando a Base de dados

```python
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
#exportar o DataFrame que criamos, e salvá-lo em um formato padrão csv.
#porem tem que ficar atento ao comando index = False, pq senão irá salvar uma tabela como novo index
```
