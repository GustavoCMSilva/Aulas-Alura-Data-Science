```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep= ';')
dados.head(1)
```

```python
# Selecione somente os imóveis classificados com tipo 'Apartamento'.
selecao = dados['tipo'] == 'Apartamento'  #vai retornar True, tudo que for apartamento
n1 = dados[selecao] #Ver a sequencia de frequencia no banco de dados

n1 #vai listar somente apartamentos
```

```python
dados[selecao].shape[0] #mostra quantas linhas tem a base de apartamentos
```

```python
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
selecao = (dados['tipo'] == 'Casa') | (dados['tipo'] == 'Casa de Condomínio') | (dados['tipo'] == 'Casa de Vila') # operador OU = |
n2 = dados[selecao] #Ver a sequencia de frequencia no banco de dados

n2 #vai mostrar tudo que for casa, casa de condominio e casa de vila
    #tem 2212 itens nessa seleçao
```

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
selecao = (dados['area'] >= 60) & (dados['area'] <= 100) #operador &
n3 = dados[selecao]

n3 # vai selecionar tudo que for - 60 <=area <= 100
```

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
selecao = (dados['quartos'] >= 4) & (dados['valor'] < 2000)
n4 = dados[selecao]

n4
```

```python
print(f'Numero de apartamentos - {n1.shape[0]}')
print(f"Numero de Casas - {n2.shape[0]}")
print(f"Numero de imoveis entre 60 e 100 m2 - {n3.shape[0]}")
print(f"Numero de imoveis 4 quarto e menor que 2mil - {n4.shape[0]}")
#para mostrar somente a quantidade de cada item pedido
```
