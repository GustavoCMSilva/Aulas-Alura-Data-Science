### Criando e excluindo novas variaveis / colunas / linhas

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')

dados
```

```python
#criar nova coluna na tabela e seus valores
dados['valor bruto'] = dados['valor'] + dados['condominio'] + dados['iptu']

dados
```

```python
#criando coluna de m2
dados['valor m2'] = dados['valor'] / dados['area']
dados['valor m2'] = dados['valor m2'].round(2)

dados
```

```python
#criar coluna considerando o valor bruto
dados['valor bruto m2'] = (dados['valor bruto'] / dados['area']).round(2)

dados
```

```python
#identificaçao de casas e apartamentos
#identificando as casas
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
                                    #apply aplica a funçao para cada registro da tabela
dados['tipo agregado'] = dados['tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
#Vai dar um apply na coluna TIPO, para tudo que for dos tipo que estiverem dentro da variavel CASA
```

```python
dados #veja na ultima coluna que agora ou é casa ou é apartamento
```

```python
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas'])

alunos
```

```python
alunos['Notas-Média(Notas)'] = alunos['Notas'].apply(lambda x: x - alunos['Notas'].mean())
    
alunos
```

```python
alunos['Faixa Etária'] = alunos['Idade'].apply(lambda x: 'Menor que 20 anos' if x < 20 
        else ('Entre 20 e 40 anos' if (x >= 20 and x <= 40) 
            else 'Maior que 40 anos'))

alunos
```

```python
alunos['Notas-Média(Notas)'] = alunos.Notas - alunos.Notas.mean()
    
alunos
```
