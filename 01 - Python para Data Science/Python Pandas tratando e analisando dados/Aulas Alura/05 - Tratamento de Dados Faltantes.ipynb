##### Em seguida basta trocar os valores do IPTU que contenham NaN por 0

```python
dados
```

```python
dados[dados['condominio'].isnull()].shape[0] #Visualizar quantos registros na coluna condominio tem NaN
#porem não dá para eliminar todos NaN de condominio pois nao faz sentido.
```

Temos as variáveis Condomínio e IPTU com alguns problemas. No caso de Condominio, os dados nulos nem sempre são incorretos, afinal no caso de uma casa, realmente não há esse tipo de tarifa. Para este caso, devemos considerar a variável Tipo: caso o imóvel seja um apartamento, os dados nulos serão excluídos.

```python
#criando uma seleçao onde contenha uma seleçao de apartamentos que o condominio seja NaN
selecao = (dados['tipo'] == 'Apartamento') & (dados['condominio'].isnull())
```

```python
#apos selecionar os apartamentos que contenham dados NaN, temos que excluir esta seleçao da lista
A = dados.shape[0]
dados = dados[~selecao] #o ~ inverte a seleçao, o que é true vira false e o que é false vira true
B = dados.shape[0]
A - B #foram eliminados 745 dados
```

```python
dados[dados['condominio'].isnull()].shape[0]
```

```python
#substituindo valores de IPTU que contenham NaN pelo valor 0
dados = dados.fillna({'condominio': 0, 'iptu': 0})
```

```python
dados.info()
```

```python
#Depois de limpar o arquivo, basta salvar o novo arquivo
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```

```python
imoveis = pd.DataFrame([['Apartamento', None, 970, 68], 
                        ['Apartamento', 2000, 878, 112], 
                        ['Casa', 5000, None, 500], 
                        ['Apartamento', None, 1010, 170], 
                        ['Apartamento', 1500, 850, None], 
                        ['Casa', None, None, None], 
                        ['Apartamento', 2000, 878, None], 
                        ['Apartamento', 1550, None, 228], 
                        ['Apartamento', 2500, 880, 195]], 
                        columns = ['Tipo', 'Valor', 'Condominio', 'IPTU'])

imoveis
```

```python
# Eliminar os registros que não apresentam a variável Valor:
imoveis.dropna(subset= ['Valor'], inplace= True)

imoveis
```

```python
# Elimina os imóveis do tipo Apartamento que não apresentam valor Condominio:
ex_sel = (imoveis['Tipo'] == 'Apartamento') & (imoveis['Condominio'].isnull())
imoveis = imoveis[~ex_sel] # atençao ao cmd ~

imoveis
```

```python
# Substitui os valores faltantes que restam nas variáveis Condominio e IPTU por zero:
imoveis = imoveis.fillna({'Condominio': 0, 'IPTU': 0})

imoveis
```

```python
#reorganiza o indice da tabela que ficou errado (1 2 4 6 8 -> 0 1 2 3 4)
imoveis.index = range(imoveis.shape[0])

imoveis
```
