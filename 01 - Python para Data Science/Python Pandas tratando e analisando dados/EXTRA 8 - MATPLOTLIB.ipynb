```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('extras/aluguel.csv', sep = ';')
dados
```

```python
#area que queremos do grafico
area = plt.figure()
```

```python
#em AREA terá 4 graficos
g1 = area.add_subplot(2, 2, 1) #este terá 2 linha e 2 colunas e vai ficar na posição 1
g2 = area.add_subplot(2, 2, 2) #este terá 2 linha e 2 colunas e vai ficar na posição 2
g3 = area.add_subplot(2, 2, 3) #este terá 2 linha e 2 colunas e vai ficar na posição 3
g4 = area.add_subplot(2, 2, 4) #este terá 2 linha e 2 colunas e vai ficar na posição 4
```

```python
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor X Area')

g2.hist(dados.Valor)
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100) #dentro da coluna valor vai pegar 100 dados aleatorios
dados_g3.index = range(dados_g3.shape[0])  #corrigindo o index que vai bagunçar devido ao cmd de cima
g3.plot(dados_g3) #sempre vai mudar os valores, pois pega valores aleatorios do bd
g3.set_title('Amostra Valor')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index #definindo os parametros do rotulo
valores = grupo.mean().values #definindo os valores
g4.bar(label, valores)
g4.set_title('Valor Medio por Tipo')

area
```

```python
area.savefig('extras/grafico.png', dpi = 300, bbox_inches = 'tight')
#salvando o grafico               qualidade      bordas
```
