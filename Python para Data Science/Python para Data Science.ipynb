```python
#receber o valor dentro da funçao
def pode_dirigir(idade):
  idade = input('Qual a idade? ') #definindo aqui o valor da variavel, converter em inteiro
  idade = int(idade) #trasformando em inteiro
  if idade >= 18:
    print('Pode dirigir')
  else:
    print('Não pode dirigir')

pode_dirigir(idade)
```

```python
distancia = input('Quantos metros quer percorrer? ')
distancia = int(distancia)
tempo = input('Quanto segundos para chegar? ')
tempo = int(tempo)

def velocidade(espaco, tempo):
  v = espaco / tempo
  print(f'Velocidade necessaria será de: {v} m/s')

velocidade(distancia, tempo)
```

FOR dentro e fora da lista

```python
#for fora da def
def pode_dirigir(anos):
  if anos >= 18:
    print(f'Com {anos}, Pode dirigir')
  else:
    print(f'Com {anos}, não pode dirigir')

for anos in idades:
  pode_dirigir(anos)
```

```python
#for dentro da def
def verifica_direcao(idades):
  for anos in idades:
    if anos >= 18:
      print(f'Com {anos}, Pode dirigir')
    else:
      print(f'Com {anos}, não pode dirigir')

verifica_direcao(idades)
```

```python
#outra maneira com dados booleanos
permissoes = []
idades = [20, 12, 41, 17]

def vai_dirigir(idades, permissoes):
  for anos in idades:
    if anos >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

vai_dirigir(idades, permissoes)

permissoes
```

```python
for liberacao in permissoes:
  if liberacao == True:  
    print('pode dirigir')
  else:
    print('não pode dirigir')
```

```python
import matplotlib.pyplot as plt

notas_matematica = ['Matemática',8,7,6,6,7,7,8,10]
notas_portugues = ['Português',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografia',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_portugues, notas_geografia]
```

```python
for nota in notas:
 x = list(range(1, 9))
 y = nota[1:]
 plt.plot(x, y, marker='o')
 plt.xlabel('Provas')
 plt.ylabel('Notas')
 plt.title(nota[0])
 plt.show()
```
