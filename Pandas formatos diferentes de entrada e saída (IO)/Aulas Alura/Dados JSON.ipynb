### Selecionando cursos para os alunos

```python
todas_matriculas = [] #definindo
x = np.random.rand(20) #distribui de forma aleatoria
prob = x / sum(x) #divide X pela soma de X
```

```python
#Buscar em cada linha sobre o aluno
#atribuindo os cursos de maneira aleatoria
for index, row in nomes.iterrows(): #funçao para devolver linha a linha
    id = row.id_aluno #qual o id do aluno
    matriculas = row.matriculas #quantas matriculas ele possui
    for i in range(matriculas): #eu quero um I que vá até a quantidade de matriculas que possuo
        mat = [id, np.random.choice(cursos.index, p = prob)] #definindo qual curso pegara
        todas_matriculas.append(mat)

matriculas = pd.DataFrame(todas_matriculas, columns = ['id_aluno', 'id_curso'])

matriculas #o id_aluno esta cadastrado no id_curso
```

```python
matriculas.groupby('id_curso').count().join(cursos['Cursos'])
#definindo quantos alunos tem em cada curso
```

```python
#renomeando coluna
matricula_por_curso = matriculas.groupby('id_curso').count().join(cursos['Cursos']).rename(columns= {'id_aluno' : 'Quant_alunos'})
```
