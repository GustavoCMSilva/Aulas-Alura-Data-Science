```python
total_linguas = tmdb['original_language'].value_counts()
```

```python
total_geral = total_linguas.sum() #soma de todas as categorias da coluna
print('Soma de todas linguas = ', total_geral)
total_en = total_linguas.loc['en'] #definindo o nome para EN
print('Quantidade em ingles =', total_en)
total_resto = total_geral - total_en
print('Outras linguas somadas =', total_resto)
#procurar somente por EN (somente por 1 categoria expecifica na coluna origianl language)
```

```python
#definindo o dataframe
dados = {
    'lingua' : ['ingles' , 'outros'],
    'total' : [total_en, total_resto]
}
dados = pd.DataFrame(dados) #criaÃ§Ã£o do dataframe
dados #para visualizar como ficou
```

```python
sns.barplot(x='lingua', y='total', data = dados)
#selecionando os dados a partir da tabela criada acima, chamada de dados, depois coloquei liguas no eixo X, e total no eixo Y
```

```python
plt.figure(figsize=(10,10))
plt.pie(dados['total'], labels = dados['lingua'])
```

```python
#comando para mostrar sÃ³ os que aparecem em ingles - tmdb.query('original_language == 'en'')
tplof = tmdb.query("original_language != 'en'").original_language.value_counts() #tudo que for diferente do que estou procurando
tplof #tudo que nÃ£o Ã© ingles
```

```python
tplo = tmdb.query("original_language != 'en'") #peguei tudo que nÃ£o Ã© ingles, != significa diferente
sns.catplot(x = 'original_language', kind='count', data = tplo) #plotar o grafico de tudo que nÃ£o Ã© ingles / no comando aspect 2 significa o dobro da largura
```

```python
sns.catplot(x = 'original_language', kind='count', data = tplo, aspect=2) #No comando aspect = n, significa multiplicar a largura
```

```python
#porem o grafico ainda esta fora de ordem
sns.catplot(
    x = 'original_language',
     kind='count', data = tplo,
      aspect=2,
       order=tplof.index ) #organizando o indice a partir da variavel criada anteriormente.
```

```python
#porem ainda nao esta na cor que quero
#porem o grafico ainda esta fora de ordem
sns.catplot(
    x = 'original_language', #de qual coluna eu quero
     kind='count', #ser uma contagem
     data = tplo, #de onde vem os dados
     aspect=2, # largura do grafico (multiplicaÃ§ao)
     palette='GnBu_r', # paleta de cores, a letra apos o _ define o lado que comeÃ§a, testei R e D
     order=tplof.index) #organizando o indice a partir da variavel criada anteriormente.) #qual palette de cor eu escolhi.
```
